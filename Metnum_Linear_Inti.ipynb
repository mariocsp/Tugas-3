```python
# gauss siedel
# MP
print('\nMetode gauss seidel\n')
matrix = [[23,4,-6,7],[2,-45,6,89],[45,78,189,-456]] # Matrix dominan diagonal dari persamaaan
tebak = [0,0,0] # angka tebakan, harus sebanyak variabel bebas yang dicari
batas = 0.000001 # batas error untuk menentukan akhir iterasi 
ralat = batas+1
k = 0

while ralat > batas:
    print(f'\niterasi ke-{k+1}\n')
    jumlah_residu = 0
    for i in range(len(matrix)):
        jumlah = 0
        for j in range(len(matrix)):
            jumlah += matrix[i][j]*tebak[j]
        residu = (1/matrix[i][i] * (matrix[i][-1]-jumlah))
        jumlah_residu += residu
        lengkap = tebak[i] + residu
        print(f'x{i+1}= {lengkap}')
        tebak[i] = lengkap
    ralat = (jumlah_residu**2)**0.5
    jumlah_residu = 0
    print(f'ralat = {ralat}\n-----------------------------')
    k += 1

tebak.clear()

```

```python
#jacobi
#MP
print('\nMetode Jacobi\n')
matrix = [[23,4,-6,7],[2,-45,6,89],[45,78,189,-456]] # Matrix dominan diagonal dari persamaaan
tebak = [0,0,0] # angka tebakan, harus sebanyak variabel bebas yang dicari
tebak_iterasi = []  # list untuk menyimpan nilai x
batas = 0.000001 # batas error untuk menentukan akhir iterasi 
ralat = batas+1
k = 0

while ralat > batas:
    print(f'\niterasi ke-{k+1}\n')
    jumlah_residu = 0
    for i in range(len(matrix)):
        jumlah = 0
        for j in range(len(matrix)):
            jumlah += matrix[i][j]*tebak[j]
        residu = (1/matrix[i][i] * (matrix[i][-1]-jumlah))
        jumlah_residu += residu
        lengkap = tebak[i] + residu
        print(f'x{i+1}= {lengkap}')
        tebak_iterasi.append(lengkap)
    tebak = tebak_iterasi.copy()
    tebak_iterasi.clear()
    ralat = (jumlah_residu**2)**0.5
    jumlah_residu = 0
    print(f'ralat = {ralat}\n-----------------------')
    k += 1
```

```python
import pandas as pd

matrix = [[23,4,-6,7],[2,-45,6,89],[45,78,189,-456]] # Matrix dominan diagonal dari persamaaan
tebak = [0,0,0] # angka tebakan, harus sebanyak variabel bebas yang dicari
batas = 0.000001 # batas error untuk menentukan akhir iterasi 
ralat = batas+1
k = 0
w = 1.1


tabel = {
    'Omega': [],
    'iterasi': [],
    'X1': [],
    'X2': [],
    'X3': [],
    'Galat Pendekatan relatif':[]
}

while ralat > batas:
    print(f'\niterasi ke-{k + 1}\n')
    jumlah_residu = 0
    for i in range(len(matrix)):
        jumlah = 0
        for j in range(len(matrix)):
            jumlah += matrix[i][j] * tebak[j]
        residu = (1 / matrix[i][i] * (matrix[i][-1] - jumlah))
        jumlah_residu += residu
        lengkap = tebak[i] + residu
        SOR = lengkap * w + tebak[i] * (1 - w)
        print(f'x{i + 1}= {SOR}')
        tebak[i] = SOR
        tabel[f'X{i + 1}'].append(round(tebak[i], 6))
    tabel['iterasi'].append(f'Iterasi ke-{k+1}')
    ralat = (jumlah_residu ** 2) ** 0.5
    tabel['Galat Pendekatan relatif'].append(f'{ralat:9G}')
    tabel['Omega'].append(w)
    jumlah_residu = 0
    print(f'ralat = {ralat}\n---------------------------')
    k += 1

def tulis():
    tabel_excel = pd.DataFrame(tabel)
    tulis = pd.ExcelWriter(r"C:\Users\dell\Documents\Output_python.xlsx")  # masukan directory file tujuam
    tabel_excel.to_excel(tulis)
    tulis.save()



```
